```python
# imports
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
```

```python
url = 'https://www.allrecipes.com/recipes/839/desserts/cookies/chocolate-chip-cookies/'
response = requests.get(url)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find the elements for recipe titles and URLs
recipe_cards = soup.find_all('a', class_='comp mntl-card-list-items mntl-document-card mntl-card card card--no-image')
```

```python
recipe_data_list = []

# Iterate over the recipe cards
for card in recipe_cards:
    print(card['href'])
    response = requests.get(card['href'])

    # Create a BeautifulSoup object
    soup = BeautifulSoup(response.text, 'html.parser')

    # get script tags
    script_tag = soup.find('script', type='application/ld+json')
    recipe_data = json.loads(script_tag.string)

    # recipe stats
    recipe_data_entry = {}

    # add recipe directions together
    recipe_directions = ""
    if 'recipeInstructions' in recipe_data[0]:
      for step in recipe_data[0]['recipeInstructions']:
        recipe_directions += step['text'] + "\n"
    recipe_data_entry["recipeDirections"]= recipe_directions
    print('recipe_directions', recipe_data_entry["recipeDirections"])

    for col in ['name', 'cookTime', "prepTime", "totalTime"]:
      if col in recipe_data[0]:
        recipe_data_entry[col] = recipe_data[0][col]
      else:
        recipe_data_entry[col] = pd.NA

    if "aggregateRating" in recipe_data[0]:
      for col in ['ratingValue', "ratingCount"]:
        if col in recipe_data[0]["aggregateRating"]:
          recipe_data_entry[col] = recipe_data[0]["aggregateRating"][col]
        else:
          recipe_data_entry[col] = pd.NA

    if "nutrition" in recipe_data[0]:
      for col in ['calories', 'carbohydrateContent', 'cholesterolContent', 'fiberContent', 'proteinContent', 'saturatedFatContent',
      'sodiumContent', 'fatContent', 'unsaturatedFatContent']:
        if col in recipe_data[0]["nutrition"]:
          recipe_data_entry[col] = recipe_data[0]["nutrition"][col]
        else:
          recipe_data_entry[col] = pd.NA

    # iterate through each ingredient and add it to the dictionary
    ingredients = ""
    if "recipeIngredient" in recipe_data[0]:
      for ingredient in recipe_data[0]["recipeIngredient"]:
          ingredients += ingredient + '\n'
          # Split the ingredient into amount and name
          split_line = ingredient.split(' ', 2)
          ingredient_amount = ' '.join(split_line[:2])
          ingredient_name = ' '.join(split_line[2:])

          if 'flour' in ingredient.lower():
            ingredient_name = 'flour' 
          elif 'brown sugar' in ingredient.lower():
            ingredient_name = 'brown sugar'
          elif 'sugar' in ingredient.lower():
            ingredient_name = 'white sugar'
          elif 'vanilla extract' in ingredient.lower():
            ingredient_name = 'vanilla extract'
          elif 'salt' in ingredient.lower():
            ingredient_name = 'salt'
          elif 'butter' in ingredient.lower():
            ingredient_name = 'butter'
          elif 'baking soda' in ingredient.lower():
            ingredient_name = 'baking soda'
          elif 'baking powder' in ingredient.lower():
            ingredient_name = 'baking powder'
          elif 'egg' in ingredient.lower():
            ingredient_name = 'eggs'
          elif 'shortening' in ingredient.lower():
            ingredient_name = 'shortening'
          elif 'chocolate' in ingredient.lower():
            if 'semisweet' in ingredient.lower():
              recipe_data_entry['chocolate type'] = 'semisweet'
            elif 'white' in ingredient.lower():
              recipe_data_entry['chocolate type'] = 'white'
            elif 'milk' in ingredient.lower():
              recipe_data_entry['chocolate type'] = 'milk'
            else:
              recipe_data_entry['chocolate type'] = 'other'
            ingredient_name = 'chocolate'
          elif 'unsalted butter' in ingredient.lower():
            ingredient_name = 'unsalted butter'
          elif 'butter' in ingredient.lower() and "peanut butter" not in ingredient.lower():
            ingredient_name = 'salted butter'
          elif 'water' in ingredient.lower():
            ingredient_name = 'water'
          elif 'oat' in ingredient.lower():
            ingredient_name = 'oats'
          elif 'walnuts' in ingredient.lower():
            ingredient_name = 'walnuts'
          elif 'nuts' in ingredient.lower() or 'pecans' in ingredient.lower() or 'pistachios' in ingredient.lower():
            ingredient_name = 'nuts'
          elif 'oil' in ingredient.lower():
            ingredient_name = 'oil'  
          elif 'cinnamon' in ingredient.lower():
            ingredient_name = 'cinnamon'
          else: 
            ingredient_name = 'other'
          recipe_data_entry[ingredient_name] = ingredient_amount
      recipe_data_entry['ingredients'] = ingredients

    # add recipe info to list
    recipe_data_list.append(recipe_data_entry)

recipe_data_list
```

```python
# dict to df
recipe_data_df = pd.DataFrame(recipe_data_list)
recipe_data_df
```

```python
# print counts for each column
for col in recipe_data_df.columns:
  print(col, recipe_data_df[col].count())
```

```python
# convert times and nutrition stats into ints
for col in ['calories', 'carbohydrateContent', 'cholesterolContent', 'fiberContent', 'proteinContent', 'saturatedFatContent',
      'sodiumContent', 'fatContent', 'unsaturatedFatContent', 'cookTime', "prepTime", "totalTime", 'eggs']:
      recipe_data_df[col] = recipe_data_df[col].str.extract('(\d+)')
```

```python
# convert baking ingredients from one unit to another (cups, tablespoons, teaspoons)
def convert_ingredients(num, output_unit):

  # Define conversion ratios
  conversion_ratios = {
        ('cup', 'tablespoon'): 16,
        ('tablespoon', 'teaspoon'): 3,
        ('cup', 'teaspoon'): 48,
        ('tablespoon', 'cup'): 1/16,
        ('teaspoon', 'tablespoon'): 1/3,
        ('teaspoon', 'cup'): 1/48,
        ('cup', 'cup'): 1,
        ('teaspoon', 'teaspoon'): 1,
        ('tablespoon', 'tablespoon'): 1
    }

  # Split the quantity string into value and unit
  try:
    num_value, input_unit = num.split()

    if 'cup' in num.lower():
      input_unit = 'cup'
    elif 'tablespoon' in num.lower():
      input_unit = 'tablespoon'
    elif 'teaspoon' in num.lower():
      input_unit = 'teaspoon'

    # Convert the value to a float
    num_value = float(num_value)

  except:
    return None

  # Check if conversion is possible
  if (input_unit, output_unit) in conversion_ratios:
      # Perform conversion
      ratio = conversion_ratios[(input_unit, output_unit)]
      result = num_value * ratio
      return result
  else:
      # Conversion is not supported
      return None
```

```python
# convert to cups
for col in ['butter', 'white sugar', 'brown sugar','water', 'flour', 'chocolate', 'walnuts',
   'oats', 'nuts', 'other', 'oil', 'shortening']:
   print(col)
   recipe_data_df[col] = recipe_data_df[col].apply(convert_ingredients, args=('cup',))
```

```python
# convert to teaspoons
for col in [ 'vanilla extract', 'baking soda', 'salt', 'baking powder', 'cinnamon']:
   print(col)
   recipe_data_df[col] = recipe_data_df[col].apply(convert_ingredients, args=('teaspoon',))
```

```python
# fill nan ingredient amounts with 0
recipe_data_df[['butter','white sugar', 'brown sugar', 'eggs', 'vanilla extract', 'baking soda',
       'water', 'salt', 'flour', 'chocolate', 'walnuts', 'oats', 'nuts', 'other', 'oil', 'baking powder',
       'cinnamon', 'shortening']] = recipe_data_df[['butter','white sugar', 'brown sugar', 'eggs', 'vanilla extract', 'baking soda',
       'water', 'salt', 'flour', 'chocolate', 'walnuts', 'oats', 'nuts', 'other', 'oil', 'baking powder',
       'cinnamon', 'shortening']].fillna(0)
recipe_data_df
```

```python
# drop rows where name is nan
recipe_data_df.dropna(subset='name', inplace=True)

# set other to true/false
recipe_data_df['other'] = recipe_data_df['other'] > 0

# if dough is chilled before baking
recipe_data_df['dough chilled'] = recipe_data_df['recipeDirections'].str.contains('chill|refr', case=False)

# extract temp from directions
recipe_data_df['temperature'] = recipe_data_df['recipeDirections'].str.extract(r'(\d+) degrees F')
recipe_data_df
```

```python
# save 
recipe_data_df[['name', 'cookTime', 'prepTime', 'totalTime',
       'ratingValue', 'ratingCount', 'calories', 'carbohydrateContent',
       'cholesterolContent', 'fiberContent', 'proteinContent',
       'saturatedFatContent', 'sodiumContent', 'fatContent',
       'unsaturatedFatContent', 'butter', 'white sugar', 'brown sugar', 'eggs',
       'vanilla extract', 'baking soda', 'water', 'salt', 'flour',
       'chocolate type', 'chocolate', 'walnuts', 'oats', 'nuts',
      'oil', 'baking powder', 'cinnamon', 'shortening', 'other',
       'temperature', 'dough chilled', 'ingredients', 'recipeDirections']].to_csv('cookies.csv')
```
